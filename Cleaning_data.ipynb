```python
import pandas as pd
```

```python
# Importando o dataframe
tech_data = pd.read_csv('C:/Users/lucas/Desktop/Projeto/TechnologyIndex.csv')
tech_data.head()
```

```python
# Limpeza de dados do dataframe
tech_data.info()
```

```python
tech_data.isna().any() # NÃ£o possue valores nulos
```

```python
# Removendo '$' de todas as linhas e colunas
tech_data.columns
colunas = tech_data[['iPhone', 'Android', 'MacBook', 'Windows Powered',
       'PS4', 'Xbox one', 'iPad mini', 'Samsung tablet', '40 inch smart TV',
       'Apple Watch', 'Brand headphone', 'hard drive 2TB', 'Portable charger',
       'Printer']]


for col in colunas:
    tech_data[col] = tech_data[col].str.replace('$', '')

tech_data
```

```python
# Removendo ',' de todas as linhas e colunas

for col in colunas:
    tech_data[col] = tech_data[col].str.replace(',', '')

tech_data
```

```python
# Corrigindo o Dtype das colunas

for col in colunas:
    tech_data[col] = tech_data[col].astype(dtype='float')

tech_data.info()
```

```python
# Salvando o banco de dados limpo
tech_data.to_csv('cleaned_data.csv', index=False)
```
